# Fundamentals of Artificial Neural Networks (ANN)

## Structure of an ANN
An Artificial Neural Network is composed of layers of interconnected nodes (neurons), each performing mathematical computations.

### Main Components:
- **Input Layer:** Receives input data.
- **Hidden Layers:** Perform computations using weights and biases.
- **Output Layer:** Provides final predictions.

## Activation Functions
Activation functions introduce non-linearity into the model, allowing the ANN to learn complex patterns.

### Common Activation Functions:
1. **Sigmoid Function:**
   $ f(x) = \frac{1}{1 + e^{-x}} $
   - Output range: (0, 1)
   - Good for probabilities but prone to vanishing gradients.

2. **ReLU (Rectified Linear Unit):**
   $ f(x) = \max(0, x) $
   - Most widely used.
   - Helps mitigate vanishing gradients.

3. **Tanh (Hyperbolic Tangent):**
   $ f(x) = \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} $
   - Output range: (-1, 1)
   - Centered around zero, useful for balanced data.

---

### Code Example: Visualizing Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
x = np.linspace(-10, 10, 100)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, tanh(x), label='Tanh')
plt.title("Activation Functions")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.legend()
plt.grid(True)
plt.show()
```
